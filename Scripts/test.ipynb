```python
df = df_ob.copy()

df = df[['Supplier PO #','Ship-to country']].drop_duplicates(subset='Supplier PO #')
df = remove_decimals(df, 'Supplier PO #')
df = df.rename(columns={'Supplier PO #': 'PO', 'Ship-to country': 'COUNTRY'})

#Array to identify destination country for inbound process, US: USA, AR: Argentina, PR: Puerto Rico, AK: Alaska
inbound_process = ['US', 'AR', 'PR', 'AK']
df['INBOUND'] = df['COUNTRY'].apply(lambda x: 'DOMESTIC' if x in inbound_process else 'INBOUND/INTERNATIONAL')

df = df[['COUNTRY','INBOUND']][~df['COUNTRY'].isin(inbound_process)].drop_duplicates(subset='COUNTRY').reset_index(drop=True)

df
```
